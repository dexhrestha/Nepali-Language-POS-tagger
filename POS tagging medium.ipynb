```python
def add_basic_features(sentence_terms,index):
    term = sentence_terms[index]
    return {
        'nb_terms' : len(sentence_terms),
        'term':term,
        'is_first' : index == 0,
        'is_last' : index == len(sentence_terms) - 1,
#         'is_capitalized' : term[0].upper() == term[0],
#         'is_all_caps' : term.upper() == term,
#         'is_all_lower': term.lower() == term,
        'prev_word': '' if index == 0 else sentence_terms[index - 1],
        'next_word': '' if index == len(sentence_terms) - 1 else sentence_terms[index + 1]
    }
```

```python
def untag(tagged_sentence):
    return [w for w ,_ in tagged_sentence]

def transform_to_dataset(tagged_sentences):
    X,y = [], []
    for pos_tags in tagged_sentences:
#         print(pos_tags)        
        for index, (term,class_) in enumerate(pos_tags):
#             print(index,term,class_)
            X.append(add_basic_features(untag(pos_tags),index))
            y.append(class_)
    return X,y
```

```python
# print(training_sentences)
X_train, y_train = transform_to_dataset(training_sentences)
X_test, y_test = transform_to_dataset(testing_sentences)
X_val, y_val = transform_to_dataset(validation_sentences)
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import LabelEncoder
import pickle 
dict_file = 'dict_file.sav'
label_file = 'label_file.sav'

dict_vectorizer = DictVectorizer(sparse=False)
dict_vectorizer.fit(X_train+X_test)
pickle.dump(dict_vectorizer,open(dict_file,'wb'))

label_encoder = LabelEncoder()
label_encoder.fit(y_train+y_test)
pickle.dump(label_encoder,open(label_file,'wb'))
```

```python
X_train = np.asarray(X_train)
X_val=np.asarray(X_val)
X_test = np.asarray(X_test)
print(X_train.shape,X_test.shape,X_val.shape)
```

```python
import math
from keras.utils import np_utils

def generator(samples,dict_vectorizer,label_encoder,batch_size):
    n_samples = len(samples)
    n_batches = math.ceil(n_samples/batch_size)
    counter = 0 
    X , y = transform_to_dataset(samples)

#     print(n_samples)
    while 1:
        X_n = dict_vectorizer.transform(X[batch_size*counter:batch_size*(counter+1)])
        # Fit LabelEncoder with our list of classes
        
        # Encode class values as integers
#         y_train = label_encoder.transform(y_train)
#         y_test = label_encoder.transform(y_test)
        y_n = label_encoder.transform(y[batch_size*counter:batch_size*(counter+1)])
        
#         y_train = np_utils.to_categorical(y_train)
#         y_test = np_utils.to_categorical(y_test)
        y_n = np_utils.to_categorical(y_n)
#         print('y',y.shape)
        counter+=1

        yield (X_n,y_n)
        
        if counter <= n_batches:
            counter = 0
```

# Bulding Model

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
```

```python
def build_model(input_dim, hidden_neurons, output_dim):
    model = Sequential(
    [
        Dense(hidden_neurons, input_dim=input_dim),
        Activation('relu'),
        Dropout(0.2),
        Dense(hidden_neurons),
        Activation('relu'),
        Dropout(0.2),
        Dense(output_dim, activation='softmax')
    ])
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python


hist = model.fit_generator(generator(training_sentences,dict_vectorizer,label_encoder,batch_size),steps_per_epoch = math.ceil(X_train.shape[0]/batch_size), epochs=5,validation_data=generator(validation_sentences,dict_vectorizer,label_encoder,batch_size),verbose=1,validation_steps=X_val.shape[0]/batch_size*2)
```

```python
import matplotlib.pyplot as plt
def plot_model_performance(train_loss, train_acc, train_val_loss, train_val_acc):
    """ Plot model loss and accuracy through epochs. """
    blue= '#34495E'
    green = '#2ECC71'
    orange = '#E23B13'
    # plot model loss
    fig, (ax1, ax2) = plt.subplots(2, figsize=(10, 8))
    ax1.plot(range(1, len(train_loss) + 1), train_loss, blue, linewidth=5, label='training')
    ax1.plot(range(1, len(train_val_loss) + 1), train_val_loss, green, linewidth=5, label='validation')
    ax1.set_xlabel('# epoch')
    ax1.set_ylabel('loss')
    ax1.tick_params('y')
    ax1.legend(loc='upper right', shadow=False)
    ax1.set_title('Model loss through #epochs', color=orange, fontweight='bold')
    # plot model accuracy
    ax2.plot(range(1, len(train_acc) + 1), train_acc, blue, linewidth=5, label='training')
    ax2.plot(range(1, len(train_val_acc) + 1), train_val_acc, green, linewidth=5, label='validation')
    ax2.set_xlabel('# epoch')
    ax2.set_ylabel('accuracy')
    ax2.tick_params('y')
    ax2.legend(loc='lower right', shadow=False)
    ax2.set_title('Model accuracy through #epochs', color=orange, fontweight='bold')
```

```python
plot_model_performance(
    train_loss=hist.history.get('loss', []),
    train_acc=hist.history.get('acc', []),
    train_val_loss=hist.history.get('val_loss', []),
    train_val_acc=hist.history.get('val_acc', [])
)
```

```python
def transform_to_input(sentence):  
    '''
    params: list containing each word in sentence,
    return: list list of dict of sentence to dict
    '''
    ip = []
#     for sentence in sentences:
    for index,word in enumerate(sentence):
        ip.append(add_basic_features(sentence,index))
    return ip


def pos_tagger(sentence):
    '''
    params: list containing each word in sentence,
    return: list sentence converted to list of (tag,word) pairs
    '''
    transformed_sentence = transform_to_input(sentence)
    tagged_sentence = []
    for word in transformed_sentence:
#         print(word)
#         print(dict_vectorizer.transform(word).shape)
        result = model.predict(dict_vectorizer.transform(word))
        encoded_tag = np.argmax(result)
        tagged_sentence.append((word['term'],list(label_encoder.inverse_transform([encoded_tag]))[0]))
    return tagged_sentence
```
